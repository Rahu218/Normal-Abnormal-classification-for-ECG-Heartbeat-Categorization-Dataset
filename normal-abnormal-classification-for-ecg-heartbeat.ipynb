# Exploratory Analysis

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
abNormalhHeartData = pd.read_csv('/kaggle/input/heartbeat/ptbdb_abnormal.csv', header=None)
normalhHeartData = pd.read_csv('/kaggle/input/heartbeat/ptbdb_normal.csv', header=None)

print('Abnormal heart data information')
abNormalhHeartData.info()

print('Normal heart data information')
normalhHeartData.info()
```

```python
abNormalhHeartData.head()
```

```python
normalhHeartData.head()
```

```python
_, (ax1,ax2) = plt.subplots(2,1, figsize=(16,9))

ax1.plot(normalhHeartData.loc[:,:186].mean(), label='noraml')
ax1.plot(abNormalhHeartData.loc[:,:186].mean(), label='abnoraml')
ax1.set_title("Mean of the data points for heart data")
ax1.legend()


ax2.plot(normalhHeartData.loc[:,:186].std(), label='noraml')
ax2.plot(abNormalhHeartData.loc[:,:186].std(), label='abnoraml')
ax2.legend()
ax2.set_title("STD for heart data")
```

**Concatenate the abnormal data with the normal one**

```python
all_heart_data = pd.concat([abNormalhHeartData, normalhHeartData], axis=0, ignore_index=True)
# all_heart_data = all_heart_data.sample(frac=1, random_state=0) # shuffle the data 
print(all_heart_data.info())
all_heart_data.head()
```

**Make sure that we have only two classes**

```python
all_heart_data.loc[:,187].value_counts()
```

**Split the data into X and y | train and test**

```python
X = all_heart_data.loc[:,:186]
y = all_heart_data.loc[:,187]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0)
```

# Train a model using LogisticRegression

```python
log_model = LogisticRegression(random_state=0)
log_model.fit(X_train, y_train)
```

**Evaluate the LogisticRegression model   
Accuracy of the model using the train and test data.**

**Take one point**

```python
predicted_point = log_model.predict([X_test.iloc[0,:]])
print('Predict value using the LogisticRegression model: ', end='')
print(predicted_point)
print('The true value: ',end='')
print(y_test.iloc[0])
```

**Accuracy on the train data**

```python
print(f'The accuracy of the LogisticRegression model on train data = {log_model.score(X_train, y_train)*100:.3}%')
```

**Accuracy on the test data**

```python
print(f'The accuracy of the LogisticRegression model on test data = {log_model.score(X_test, y_test)*100:.3}%')
```

**Confusion Matrix**

```python
predicted = log_model.predict(X_test)
actual =  y_test
confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

# Train a model using DecisionTreeClassifier

```python
dec_tree_model = DecisionTreeClassifier(random_state=0)
dec_tree_model.fit(X_train, y_train)
```

**Accuracy on the train data**

```python
print(f'The accuracy of the DecisionTreeClassifier model on train data = {dec_tree_model.score(X_train,y_train)*100:.5}%')
```

**Confusion Matrix**

```python
predicted = dec_tree_model.predict(X_test)
actual =  y_test
confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

# Train a model using KNeighborsClassifier

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
```

**Accuracy on the train data**

```python
print(f'The accuracy of the knn model on train data = {knn_model.score(X_train,y_train)*100:.3}%')
```

**Accuracy on the test data**

```python
print(f'The accuracy of the knn model on test data = {knn_model.score(X_test,y_test)*100:.3}%')
```

**Confusion Matrix**

```python
predicted = knn_model.predict(X_test)
actual =  y_test
confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

# Train a model using RandomForestClassifier

```python
rand_model = RandomForestClassifier()
rand_model.fit(X_train, y_train)
```

**Accuracy on the train data**

```python
print(f'The accuracy of the RandomForest model on train data = {rand_model.score(X_train,y_train)*100:.5}%')
```

**Accuracy on the test data**

```python
print(f'The accuracy of the RandomForest model on test data = {rand_model.score(X_test,y_test)*100:.3}%')
```

```python
predicted = rand_model.predict(X_test)
actual =  y_test
confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```
